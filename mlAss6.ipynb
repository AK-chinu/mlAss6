```python
import numpy as np
from scipy.stats import pearsonr, spearmanr

# Sample data for number of hours of exercise per week and BMI
exercise_hours = np.array([4, 5, 3, 6, 2, 5, ...])  # Insert your data
bmi = np.array([25, 27, 24, 29, 26, 28, ...])  # Insert your data

# Calculate Pearson correlation coefficient
pearson_corr, _ = pearsonr(exercise_hours, bmi)

# Calculate Spearman's rank correlation coefficient
spearman_corr, _ = spearmanr(exercise_hours, bmi)

print("Pearson Correlation Coefficient:", pearson_corr)
print("Spearman's Rank Correlation Coefficient:", spearman_corr)
```

```python
import numpy as np

# Sample data for number of hours individuals spend watching television per day and level of physical activity
tv_hours = np.array([3, 2, 4, 3, 5, ...])  # Insert your data
physical_activity = np.array([2, 3, 1, 2, 1, ...])  # Insert your data

# Calculate mean of tv_hours and physical_activity
mean_tv_hours = np.mean(tv_hours)
mean_physical_activity = np.mean(physical_activity)

# Compute covariance between tv_hours and physical_activity
covariance = np.cov(tv_hours, physical_activity)[0, 1]

# Compute standard deviations of tv_hours and physical_activity
std_tv_hours = np.std(tv_hours)
std_physical_activity = np.std(physical_activity)

# Calculate Pearson correlation coefficient
pearson_corr = covariance / (std_tv_hours * std_physical_activity)

print("Pearson Correlation Coefficient:", pearson_corr)
```
